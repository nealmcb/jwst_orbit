# Plot Web Space Telescope positions from JPL Horizons

```python
from astropy import time
from poliastro.ephem import Ephem
from poliastro.bodies import Earth, Moon, Sun
from poliastro.plotting import OrbitPlotter3D
```

```python
day = time.TimeDelta(1)
```

Get positions for first 31 days (all that Horizons provides now)

```python
jwst_epoch = time.Time("2021-12-25 12:50")  # UTC by default
```

```python
epochs = time.Time([jwst_epoch + interval * day for interval in range(31)])
```

```python
jwst = Ephem.from_horizons("JWST", epochs, attractor=Earth)
```

```python
frame = OrbitPlotter3D()
```

```python
frame.set_attractor(Earth)
```

```python
frame.plot_ephem(jwst)
```
